```python
df['y'] = df['y'].map({'no': 0, 'yes': 1})

distinct_values = {column: df[column].unique() for column in df.select_dtypes(include='object').columns}


column_mappings = {column: {value: index for index, value in enumerate(distinct_values[column])} for column in distinct_values}


for column, mapping in column_mappings.items():
    df[column] = df[column].map(mapping)
    
df.head()

print(df.dtypes)
```

# Feature Selection

1. On this step, feature is selected for which machine learning model would be tranined 
2. There are several ways to find best features for model traning 
3. We are selecting features based on static threshold value with correlated columns 

```python
def plot_corelation(df):
    corr_matrix = df.corr()

    fig, ax = plt.subplots(figsize=(12, 10))

    sns.heatmap(corr_matrix, cmap="coolwarm", annot=True, fmt=".2f", linewidths=.5, ax=ax)

    plt.title("Correlation Heatmap")

    plt.show()
```

```python
all_columns = np.array(df.columns)
condition = all_columns != 'y'
sns.pairplot(df, x_vars=['y'], y_vars=all_columns[condition])
plt.show()
```

```python
df.corr()['y'].sort_values()
```

```python
plot_corelation(df)
```

```python
correlation_threshold = 0.10
correlated_columns = df.corr().abs()['y'] > correlation_threshold
# correlated_columns = df.corr()['y'] > correlation_threshold
df = df[correlated_columns.index[correlated_columns]]

print(df.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler, EditedNearestNeighbours, NearMiss
```

```python
def plot_confusion_matrix(y_test,y_pred):
    conf_matrix = confusion_matrix(y_test, y_pred)
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='coolwarm')
    plt.ylabel('Actual Label')
    plt.xlabel('Predicted Label')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
print(y_train.value_counts(normalize=True))
print(y_test.value_counts(normalize=True))
```

```python
pipeline = make_pipeline(
    ColumnTransformer(
        transformers=[
            ('numeric', StandardScaler(), X.columns)
        ],
        remainder='passthrough'
    ),
    RandomForestClassifier(random_state=42, n_estimators=100, max_depth=10)
)


pipeline.fit(X_train, y_train)


y_pred = pipeline.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", classification_rep)
print("Confusion Matrix:\n", conf_matrix)

cv_scores = cross_val_score(pipeline, X_train, y_train, cv=5)
rand_cv_scores = cv_scores 
print("Cross-validation scores:", cv_scores)
print("Mean cross-validation score:", cv_scores.mean())

plot_confusion_matrix(y_test, y_pred)
```

```python
from sklearn.svm import SVC

svm_pipeline = make_pipeline(
    ColumnTransformer(
        transformers=[
            ('numeric', StandardScaler(), X.columns)
        ],
        remainder='passthrough'
    ),
    SVC(random_state=42) 
)

svm_pipeline.fit(X_train, y_train)

y_svm_pred = svm_pipeline.predict(X_test)

svm_accuracy = accuracy_score(y_test, y_svm_pred)
svm_classification_rep = classification_report(y_test, y_svm_pred)
svm_conf_matrix = confusion_matrix(y_test, y_svm_pred)

print("SVM Accuracy:", svm_accuracy)
print("SVM Classification Report:\n", svm_classification_rep)
print("SVM Confusion Matrix:\n", svm_conf_matrix)

svm_cv_scores = cross_val_score(svm_pipeline, X_train, y_train, cv=5)
print("SVM Cross-validation scores:", svm_cv_scores)
print("Mean SVM Cross-validation score:", svm_cv_scores.mean())

plot_confusion_matrix(y_test, y_pred)
```
